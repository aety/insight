```python
from bs4 import BeautifulSoup
import requests

# URL 
# RESPONSE
# CONTENT
url = 'https://www.who.int/csr/don/archive/year/2019/en/'
response = requests.get(url, timeout=10)
content = BeautifulSoup(response.content,"html.parser")

diseaseArr = []
li = content.find('ul', attrs={"class": "auto_archive"})
for disease_date in li.findAll('li'):
    date = disease_date.find("a", href=True)
    name = disease_date.find('span', attrs={"class": "link_info"})
    #print(name.text.encode('utf-8'))
    #print(date.text)
    #print(date['href'])
    disease = {
        "name": name.text,
        "date": date.text,
        "link": date['href']
    }
    #print(disease)
    diseaseArr.append(disease)

    
#print(diseaseArr)

with open('whoDiseaseOutbreaks.json', 'w') as outfile:
    json.dump(diseaseArr, outfile)
    
#for link in soup.findAll('a', attrs={'href': re.compile("^http://")}):
#    print link.get('href')
    
#ul = content.find('ul', attrs={"class": "auto_archive"})
#li = ul.find('li')
#date = li.find("a" , href=True).text
#print(date)

#for disease_name in content.findAll('span', attrs={"class": "link_info"}):
#    print(disease_name.text)
```
