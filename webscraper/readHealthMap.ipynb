```python
from bs4 import BeautifulSoup
import requests
import json

url = 'https://www.contagionlive.com/outbreak-monitor'
response = requests.get(url, timeout=10)
content = BeautifulSoup(response.content,"html.parser")
    
outbreakArr = []

#instance = content.find('tr', attrs={"class": "outbreakInstance"})
for instance in content.findAll('tr', attrs={"class": "outbreakInstance"}):
    category = instance.find('td', attrs={"class": "outbreakCategory"})
    disease =  instance.find('td', attrs={"class": "outbreakDisease"})
    name =  instance.find('td', attrs={"class": "outbreakName"})
    location =  instance.find('td', attrs={"class": "outbreakLocation"})
    cases =  instance.find('td', attrs={"class": "outbreakCases"})
    deaths =  instance.find('td', attrs={"class": "outbreakDeaths"})
    source =  instance.find('td', attrs={"class": "outbreakSource"})
    temp = source.find("a", href=True)
    sourceUrl = temp['href']
    
    outbreak = {
        "category": category.text,
        "disease": disease.text,
        "name": name.text,
        "location": location.text,
        "cases": cases.text,
        "deaths": deaths.text,
        "sourceUrl": sourceUrl
    }
    outbreakArr.append(outbreak)    

with open('readHealthMap.json', 'w') as outfile:
    json.dump(outbreakArr, outfile)
```
