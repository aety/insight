```python
# I want to show my users live top and breaking news headlines
# Let's make a request to get live top headlines in the US right now. We'll use the /top-headlines endpoint for this.

import requests
url = ('https://newsapi.org/v2/top-headlines?'
       'country=fr&'
       'apiKey=79a135e137a24381804e7d72de05bb2d')
response = requests.get(url)
print(response.json())
```

```python
# If you want headlines just from a specific source, for example BBC News
import requests
url = ('https://newsapi.org/v2/top-headlines?'
       'sources=bbc-news&'
       'apiKey=79a135e137a24381804e7d72de05bb2d')
response = requests.get(url)
print(response.json())
```

```python
# I want to search for news articles that mention a specific topic or keyword (6 months)
import requests

url = ('https://newsapi.org/v2/everything?'
       'q=strikes&'
       'from=2019-09-10&'
       'sortBy=popularity&'
       'apiKey=79a135e137a24381804e7d72de05bb2d')

response = requests.get(url)

print(response.json())
```
