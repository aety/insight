```python
import pandas as pd
#forecast = pd.read_pickle('ground/ground_PM10_hourly_forecast.pkl')
#m = pd.read_pickle('ground/ground_PM10_hourly_m.pkl')
forecast = pd.read_pickle('franklin-d-roosevelt/PM10_hourly_forecast.pkl')
m = pd.read_pickle('franklin-d-roosevelt/PM10_hourly_m.pkl')
fname = 'metro'
pd.plotting.register_matplotlib_converters() # this resolves some issues with data type
```

```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.pyplot as plt
##pd.plotting.register_matplotlib_converters() # this resolves some issues with data type
plt.style.use('dark_background')
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot()
 
font = {'size': 20,             # font setting
       'weight':'normal'}       # font setting
matplotlib.rc('font', **font)   # font setting

fcst = forecast
fcst_t = fcst['ds'].dt.to_pydatetime()
plt.plot(m.history['ds'].dt.to_pydatetime(), m.history['y'], '.',color='#009999')
plt.plot(fcst_t, fcst['yhat'],'-', c='w')
xlabel = 'time'
ylabel = 'PM10'
plt.title(fname)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.savefig('plot_m_and_forecast'+fname+'.png',transparent=True)
```
