```python
# trying to read json using panda
from sqlalchemy import create_engine
from sqlalchemy_utils import database_exists, create_database
import psycopg2
import pandas as pd

# Define a database name 
# Set your postgres username/password, and connection specifics
username = 'postgres'
password = 'yanggnay'     # change this
host     = 'localhost'
port     = '5432'            # default port that postgres listens on
db_name  = 'auber_pm10_db'

## 'engine' is a connection to a database
## Here, we're using postgres, but sqlalchemy can connect to other things too.
engine = create_engine( 'postgresql://{}:{}@{}:{}/{}'.format(username, password, host, port, db_name) )
print(engine.url)

## create a database (if it doesn't exist)
if not database_exists(engine.url):
    create_database(engine.url)
print(database_exists(engine.url))
```

```python
# read a database from CSV and load it into a pandas dataframe
pm10_data = pd.read_pickle('auber/PM10_forecast.pkl')
```

```python
## insert data into database from Python (proof of concept - this won't be useful for big data, of course)
pm10_data.to_sql('auber_pm10_table', engine, if_exists='replace')
```

```python
# Connect to make queries using psycopg2
con = None
con = psycopg2.connect(database = 'auber_pm10_db', user = 'postgres', password = 'yanggnay', port = '5432', host = 'localhost')

# query:
sql_query = """
SELECT * FROM auber_PM10_table WHERE ds>'2019-12-30 00:00:00';
"""
pm10_data_from_sql = pd.read_sql_query(sql_query,con)
pm10_data_from_sql.head()
```
