```python
# load JSON
import json
import pandas as pd
 
data = pd.read_pickle('franklin-d-roosevelt/airQualityDataframe-franklin-d-roosevelt-PM10.pkl')
data = data.sort_values(by=['ds'])

df = data[0:100]

import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters() # this resolves some issues with data type
plt.plot(df['ds'],df['y'])
plt.show()
```

```python
# correlation between observation and lag terms 
from pandas.plotting import autocorrelation_plot
from datetime import datetime

timeStampArr = []
temp = df['ds']
for i in temp:
    test = i
    test = datetime.timestamp(test)
    #print(test)
    timeStampArr.append(test)
    

d = {'ds': timeStampArr,'y': df['y']}
dfnew = pd.DataFrame(data=d)
plt.scatter(dfnew['ds'],dfnew['y'])
```

```python
autocorrelation_plot(dfnew)
plt.show()
```
