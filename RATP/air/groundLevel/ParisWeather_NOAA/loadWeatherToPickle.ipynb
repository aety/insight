```python
import pandas as pd
df = pd.read_csv('1874183.csv')   # load data
df.TAVG = (df.TAVG-32)*5/9        # convert from F to C
df.DATE = pd.to_datetime(df.DATE) # convert to timestamp
#df.rename(columns={"DATE": "date", "TAVG": "tavg", "PRCP": "prcp"})
df.columns = map(str.lower, df.columns)
df = df[['date','tavg','prcp']]
df.head()

df.to_pickle("weatherDataframe.pkl")
```

```python
import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters() # this resolves some issues with data type
plt.plot(df.date,df.tavg)
```
