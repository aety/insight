```python
# load JSON
import json
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# hourly data 
data = pd.read_pickle('franklin-d-roosevelt/airQualityDataframe-franklin-d-roosevelt-PM10.pkl')
data = data.sort_values(by=['ds'])
```

```python
# group by year
data['grouped'] = data['ds'].dt.year
df_h = data
data_d = data.groupby(['grouped'],squeeze=True).mean()
data_d.columns = ['_'.join(col).strip() for col in data_d.columns.values]
data_d = data_d.reset_index()
data_d.columns = ['ds','y'] # rename column
df_d = data_d

data_d = data.groupby(['grouped'],squeeze=True).std()
data_d.columns = ['_'.join(col).strip() for col in data_d.columns.values]
data_d = data_d.reset_index()
data_d.columns = ['ds','y'] # rename column
df_e = data_d

plt.errorbar(df_d.ds, df_d.y, yerr=df_e.y, fmt='o-k');
print(df_d.y)
```
