```python
# load JSON
import json
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters() # this resolves some issues with data type

plt.style.use('dark_background')
font = {'size': 16,             # font setting
       'weight':'normal'}       # font setting
matplotlib.rc('font', **font)   # font setting

# hourly data 
data = pd.read_pickle('franklin-d-roosevelt/airQualityDataframe-franklin-d-roosevelt-PM10.pkl')
data = data.sort_values(by=['ds'])
```

```python
# group by year
data['grouped'] = data['ds'].dt.year
df_h = data
data_d = data.groupby(['grouped'],squeeze=True).mean()
data_d.columns = ['_'.join(col).strip() for col in data_d.columns.values]
data_d = data_d.reset_index()
data_d.columns = ['ds','y'] # rename column
df_d = data_d

data_d = data.groupby(['grouped'],squeeze=True).std()
data_d.columns = ['_'.join(col).strip() for col in data_d.columns.values]
data_d = data_d.reset_index()
data_d.columns = ['ds','y'] # rename column
df_e = data_d

plt.figure(figsize=(3,3))
plt.errorbar(df_d.ds, df_d.y, yerr=df_e.y, fmt='o-w');
plt.ylabel('PM10 $(\u03BCg/m^3)$')
plt.xticks([2013,2016,2019])
plt.yticks([0,54])
plt.title('year')
plt.savefig('historySummaries_year.png',transparent=True)
```
